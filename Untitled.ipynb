```python
import matplotlib.pyplot as plt
from __future__ import print_function
from ipywidgets import interact
import ipywidgets as widgets
from skimage.io import imread
import numpy as np

image = imread("Kropka.jpg", as_gray=True)
plt.figure(figsize=(10, 10))
plt.title("Original");
plt.imshow(image, cmap=plt.cm.Greys_r)


plt.show()
```

```python
lines = []
sin = []
width = image.shape[0]
height = image.shape[1]

def _positions(angle, offset, radius):
    x1, y1 = radius * np.cos(angle), radius * np.sin(angle)
    x2, y2 = -x1, -y1
    xv, yv = offset * np.cos(angle + np.pi / 2), offset * np.sin(angle + np.pi / 2)
    x1 += xv + radius
    y1 += yv + radius
    x2 += xv + radius
    y2 += yv + radius
    return [int(x1), int(y1), int(x2), int(y2)]

def bresenham(x1,y1,x2, y2, dx, dy, decide):
    result = []
    resultx = []
    resulty = []
    pk = 2 * dy - dx;
    for x in range(dx+1):
        if decide == 0:
            result.append((x1,y1))
            resultx.append(x1)
            resulty.append(y1)
        else:
            result.append((y1,x1))
            resultx.append(y1)
            resulty.append(x1)
        if x1 < x2:
                x1 += 1
        else:
                x1 -= 1
        if pk < 0:
            if decide == 0:
                pk = pk + 2 * dy
            else:
                pk = pk + 2 * dy
        else:
            if y1 < y2:
                y1 += 1
            else:
                y1 -= 1
            pk = pk + 2 * dy - 2 * dx
    return [result, resultx, resulty]

def _calculate_single_line(width, line, image):
        result = []
        for x, y in line:
            if 0 <= x < width and 0 <= y < width:
                result.append(image[int(x), int(y)])

        return np.array(result).mean() if len(result) > 0 else 0

def sinogram(number_of_emiters, angle):
    angles = np.arange(0, 361, angle)
    offsets = np.linspace(-width/2, width/2, number_of_emiters).astype(int)
    radius = width/2
    lines.clear()
    sin.clear()
    
    for angle in angles:
        angle = np.deg2rad(angle)
        sin_row = []
        lines_row = []
        for offset in offsets:
            positions = _positions(angle, offset, radius)
            x1 = positions[0]
            y1 = positions[1]
            x2 = positions[2]
            y2 = positions[3]
            dx = abs(x2 - x1)
            dy = abs(y2 - y1)
            if dx > dy:     
                line = bresenham(x1,y1,x2,y2, dx, dy, 0)
            else:
                line = bresenham(y1,x1,y2,x2, dy, dx, 1)
            color = _calculate_single_line(width, line[0], image)
            sin_row.append(color)
            lines_row.append(line[0])
        sin.append(sin_row)
        lines.append(lines_row)

    
    plt.figure(figsize=(10, 10))
    plt.title("Sinogram")
    plt.imshow(sin, cmap=plt.cm.Greys_r)
    plt.show()
    
    reverse = np.zeros((width, height))
    sinogram = np.array(sin)
    kernel = []


    for i in range(sinogram.shape[0]):
        for j in range(sinogram.shape[1]):
            for x, y in lines[i][j]:
                if 0 <= x < width and 0 <= y < height:
                    reverse[x, y] += sinogram[i][j]
    plt.figure(figsize=(10, 10))
    plt.title("Reverse")
    plt.imshow(reverse, cmap=plt.cm.Greys_r)

    for i in range(-10,11):
        if i == 0:
            kernel.append(1)
        elif i%2 == 0:
            kernel.append(0)
        else:
            kernel.append((-4/np.pi**2)/i**2)

    for i in range(sinogram.shape[0]):
        sinogram[i] = np.convolve(sinogram[i, :], kernel, mode='same')

    plt.figure(figsize=(10, 10))
    plt.title("Filtered sinogram")
    plt.imshow(sinogram, cmap=plt.cm.Greys_r)

    for i in range(sinogram.shape[0]):
        for j in range(sinogram.shape[1]):
            for x, y in lines[i][j]:
                if 0 <= x < width and 0 <= y < height:
                    reverse[x, y] += sinogram[i][j]
    plt.figure(figsize=(10, 10))
    plt.title("Filtered Reverse")
    plt.imshow(reverse, cmap=plt.cm.Greys_r)
interact(sinogram, number_of_emiters=widgets.IntSlider(min=1, max=400, step=1, value=180, continuous_update=False), angle=widgets.IntSlider(min=1, max=90, step=1, value=4, continuous_update=False));
```
